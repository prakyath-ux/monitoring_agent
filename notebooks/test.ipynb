# EDA to extract import key/value and features in JSONL


```python
import json
import pandas as pd
from IPython.display import display, HTML

with open("a65ee331-54af-4b95-806f-7756b3e37a9c.jsonl") as f:
    raw = [json.loads(line) for line in f]

print(f"Total entries: {len(raw)}")
print(f"Message types: {pd.Series([m['type'] for m in raw]).value_counts().to_dict()}")
```

## Split into DataFrames by type (skip queue-operation, file-history-snapshot, progress â€” not useful for prompt capturing):

```python
users = [m for m in raw if m['type'] == 'user']
assistants = [m for m in raw if m['type'] == 'assistant']
systems = [m for m in raw if m['type'] == 'system']

print(f"User messages: {len(users)}")
print(f"Assistant messages: {len(assistants)}")
print(f"System messages: {len(systems)}")
```

# User messages: extract the fields we care about:

```python
user_rows = []
for m in users:
    content_blocks = m.get('message', {}).get('content', [])
    
    # Separate actual user text from IDE/system tags
    user_text = []
    has_ide_context = False
    has_tool_result = False
    
    for block in content_blocks:
        if not isinstance(block, dict):
            continue
        if block.get('type') == 'tool_result':
            has_tool_result = True
        if block.get('type') != 'text':
            continue
        text = block['text']
        if text.startswith('<ide_opened_file>') or text.startswith('<ide_selection>'):
            has_ide_context = True
        elif text.startswith('<system-reminder>'):
            pass  # skip system noise
        elif text.strip().startswith('[Request interrupted'):
            pass  # skip interruptions
        else:
            user_text.append(text.strip())
    
    user_rows.append({
        'timestamp': m.get('timestamp', ''),
        'uuid': m.get('uuid', ''),
        'parentUuid': m.get('parentUuid'),
        'branch': m.get('gitBranch', ''),
        'prompt': ' | '.join(user_text) if user_text else None,
        'has_ide_context': has_ide_context,
        'has_tool_result': has_tool_result,
        'is_tool_response': 'toolUseResult' in m,
        'content_blocks': len(content_blocks),
    })

df_user = pd.DataFrame(user_rows)
df_user['timestamp'] = pd.to_datetime(df_user['timestamp'])
print(df_user.shape)
df_user.head(10)

        
```

##  How many user messages have actual prompts vs just IDE context:

```python
print("User message breakdown: ")
print(f"   Total:             {len(df_user)}")
print(f"   With prompt text:  {df_user['prompt'].notna().sum()}")
print(f"   Tool Responses:   {df_user['is_tool_response'].sum()}")
print(f"   With IDE context:  {df_user['has_ide_context'].sum()}")
print(f"   Empty (no text):   {(df_user['prompt'].isna() & ~df_user['is_tool_response']).sum()}")
```

## View all actual user prompts (the ones that matter):




```python
prompts = df_user[df_user['prompt'].notna()][['timestamp', 'branch', 'uuid', 'prompt']].copy()
prompts['prompt_short'] = prompts['prompt'].str[:150]
print(f"Total user prompts: {len(prompts)}")
display(HTML(prompts[['timestamp', 'branch', 'uuid','prompt_short']].to_html(index=False)))
```

## Assistant messages: extract model, tools used, files edited:

```python
asst_rows = []
for m in assistants:
    msg = m.get('message', {})
    content = msg.get('content', [])

    tools_used = []
    files_edited = []
    has_thinking = False
    text_output = []

    for block in content:
        if not isinstance(block, dict):
            continue
        if block.get('type') == 'thinking':
            has_thinking = True
        elif block.get('type') == 'text':
            text_output.append(block.get('text', '')[:200])
        elif block.get('type') == 'tool_use':
            tool_name = block.get('name', '')
            tools_used.append(tool_name)
            if tool_name in ('Edit', 'Write'):
                fp = block.get('input', {}).get('file_path', '')
                if fp:
                    files_edited.append(fp.split('/')[-1])
    
    asst_rows.append({
        'timestamp': m.get('timestamp', ''),
        'uuid': m.get('uuid', ''),
        'parentUuid': m.get('parentUuid', ''),
        'branch': m.get('gitBranch', ''),
        'model': msg.get('model', ''),
        'tools_used': tools_used,
        'tool_count': len(tools_used),
        'files_edited': list(set(files_edited)),
        'has_code_edit': len(files_edited) > 0,
        'has_thinking': has_thinking,
        'text_preview': (text_output[0][:100] if text_output else None),
    })

df_asst = pd.DataFrame(asst_rows)
df_asst['timestamp'] = pd.to_datetime(df_asst['timestamp'])
print(df_asst.shape)
df_asst

```

# Code edit analysis

```python
edits = df_asst[df_asst['has_code_edit']]
print(f"Assistant messages with code edits: {len(edits)}")
print(f"\nFiles edited (frequemcy):")
from collections import Counter
all_files = [f for files in edits['files_edited'] for f in files]
for f, c in Counter(all_files).most_common():
    print(f" {c:>4} {f}")
```

## Build the conversation chain (this is the key for prompt capturing):

```python
# Build UUID lookup for all messages
by_uuid = {}
for m in raw:
    uid = m.get('uuid', '')
    if uid:
        by_uuid[uid] = m

def find_triggering_prompt(assistant_uuid):
    """Walk parentUuid chain backward to find nearest user prompt"""
    visited = set()
    current = by_uuid.get(assistant_uuid, {}).get('parentUuid', '')
    
    while current and current not in visited:
        visited.add(current)
        node = by_uuid.get(current)
        if not node:
            return None
        
        if node.get('type') == 'user':
            content = node.get('message', {}).get('content', [])
            for block in content:
                if isinstance(block, dict) and block.get('type') == 'text':
                    text = block['text']
                    if not text.startswith('<') and len(text.strip()) > 3:
                        return text.strip()[:300]
            # This user msg had no clean text, keep walking
        
        current = node.get('parentUuid', '')
    return None

# Link every code edit to its triggering prompt
edit_prompt_pairs = []
for _, row in edits.iterrows():
    prompt = find_triggering_prompt(row['uuid'])
    edit_prompt_pairs.append({
        'timestamp': row['timestamp'],
        'files_edited': ', '.join(row['files_edited']),
        'prompt': prompt,
    })

df_pairs = pd.DataFrame(edit_prompt_pairs)
print(f"Code edits with prompt found: {df_pairs['prompt'].notna().sum()} / {len(df_pairs)}")
print(f"Code edits with NO prompt:    {df_pairs['prompt'].isna().sum()} / {len(df_pairs)}")
display(HTML(df_pairs[df_pairs['prompt'].notna()][['timestamp', 'files_edited', 'prompt']].to_html(index=False)))
```

## Tool usage summary:

```python
from collections import Counter
all_tools = [t for tools in df_asst['tools_used'] for t in tools]
tool_freq = Counter(all_tools)
print("Tool usage frequency:")
for t, c in tool_freq.most_common():
    print(f"  {c:>4}  {t}")
```
